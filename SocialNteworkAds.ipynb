```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data = data.drop('User ID', axis=1)
display(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to the 'Gender' column
data['Gender'] = le.fit_transform(data['Gender'])

display(data.head())
```

```python
X = data.drop('Purchased', axis=1)
y = data['Purchased']
# The line 'data['purchased']=y' is removed as it was creating a duplicate column.
```

```python
train_X,test_X,train_y,test_y=train_test_split(data.drop('Purchased',axis=1),data['Purchased'],test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical features for scaling (assuming 'Gender' is already encoded)
numerical_features = ['Age', 'EstimatedSalary']

# Initialize StandardScaler
scaler = StandardScaler()

# Fit the scaler on the numerical features of the training data and transform train_X
train_X[numerical_features] = scaler.fit_transform(train_X[numerical_features])

# Transform test_X using the scaler fitted on train_X
test_X[numerical_features] = scaler.transform(test_X[numerical_features])

print("Scaled train_X head:")
display(train_X.head())

print("\nScaled test_X head:")
display(test_X.head())
```

```python
knn_sklearn=KNeighborsClassifier(n_neighbors=7)
knn_sklearn.fit(train_X,train_y)
y_pred_sklearn=knn_sklearn.predict(test_X)
```

```python
sklearn_accuracy=accuracy_score(test_y,y_pred_sklearn)
print(f"Accuracy with sklearn: {sklearn_accuracy}")
sklearn_precision=precision_score(test_y,y_pred_sklearn)
print(f"Precision with sklearn: {sklearn_precision}")
sklearn_recall=recall_score(test_y,y_pred_sklearn)
print(f"Recall with sklearn: {sklearn_recall}")
sklearn_f1=f1_score(test_y,y_pred_sklearn)
print(f"F1 Score with sklearn: {sklearn_f1}")
```

```python
model=LogisticRegression()
model.fit(train_X,train_y)
y_pred=model.predict(test_X)


accuracy=accuracy_score(test_y,y_pred)
precision=precision_score(test_y,y_pred)
recall=recall_score(test_y,y_pred)
f1=f1_score(test_y,y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
from sklearn.svm import SVC

svmclf =SVC(kernel='linear')
svmclf.fit(train_X,train_y)
y_pred_svml=svmclf.predict(test_X)
```

```python
sklearn_accuracy=accuracy_score(test_y,y_pred_svml)
print(f"Accuracy with sklearn: {sklearn_accuracy}")
sklearn_precision=precision_score(test_y,y_pred_svml)
print(f"Precision with sklearn: {sklearn_precision}")
sklearn_recall=recall_score(test_y,y_pred_svml)
print(f"Recall with sklearn: {sklearn_recall}")
sklearn_f1=f1_score(test_y,y_pred_svml)
print(f"F1 Score with sklearn: {sklearn_f1}")
```

```python
svmclf =SVC(kernel='poly')
svmclf.fit(train_X,train_y)
y_pred_svmp=svmclf.predict(test_X)
```

```python
sklearn_accuracy=accuracy_score(test_y,y_pred_svmp)
print(f"Accuracy with sklearn: {sklearn_accuracy}")
sklearn_precision=precision_score(test_y,y_pred_svmp)
print(f"Precision with sklearn: {sklearn_precision}")
sklearn_recall=recall_score(test_y,y_pred_svmp)
print(f"Recall with sklearn: {sklearn_recall}")
sklearn_f1=f1_score(test_y,y_pred_svmp)
print(f"F1 Score with sklearn: {sklearn_f1}")
```

```python
sklearn_accuracy=accuracy_score(test_y,y_pred_svmpr)
print(f"Accuracy with sklearn: {sklearn_accuracy}")
sklearn_precision=precision_score(test_y,y_pred_svmpr)
print(f"Precision with sklearn: {sklearn_precision}")
sklearn_recall=recall_score(test_y,y_pred_svmpr)
print(f"Recall with sklearn: {sklearn_recall}")
sklearn_f1=f1_score(test_y,y_pred_svmpr)
print(f"F1 Score with sklearn: {sklearn_f1}")
```
