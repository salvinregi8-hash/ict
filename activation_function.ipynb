```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def linear(x):
    return x

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python

x = np.linspace(-10, 10, 100)


y_sigmoid = sigmoid(x)
y_linear = linear(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_softmax = softmax(x)


plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, y_linear, label='Linear', color='orange')
plt.title('Linear')
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, y_relu, label='ReLU', color='green')
plt.title('ReLU')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, y_tanh, label='Tanh', color='red')
plt.title('Tanh')
plt.grid(True)

plt.subplot(2, 3, 5)
x_softmax = np.linspace(-5, 5, 10)
y_softmax = softmax(x_softmax)

plt.bar(x_softmax, y_softmax)
plt.title("Softmax Activation Function")
plt.xlabel("x")
plt.ylabel("Softmax(x)")
plt.grid(True)
plt.show()
```
